```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
# takes a single argument and returns three variables (the equivalent
# matrix I, the number of rows in the matrix L, and the number of columns C).
def OpenImage(image_path):
    I = cv2.imread(image_path, cv2.IMREAD_GRAYSCALE)  # Read as grayscale
    L, C = I.shape  # Get dimensions
    return I, L, C
```

```python
import cv2
import matplotlib.pyplot as plt

def Divide(image_path):
    """
    Reads an image in RGB format and splits it into three separate channels (Red, Green, Blue).

    Parameters:
    - image_path (str): Path to the image file.

    Returns:
    - tuple: (R, G, B) channels as separate matrices.
    """
    I = cv2.imread(image_path)  # Read the image in BGR format (default for OpenCV)

    if I is None:
        raise ValueError("Error: Unable to load the image. Check the file path.")

    I = cv2.cvtColor(I, cv2.COLOR_BGR2RGB)  # Convert BGR to RGB
    R, G, B = cv2.split(I)  # Split into three separate channels

    # Display the images
    plt.figure(figsize=(10, 5))

    plt.subplot(1, 4, 1)
    plt.imshow(I)
    plt.title("Original RGB Image")

    plt.subplot(1, 4, 2)
    plt.imshow(R, cmap='Reds')
    plt.title("Red Channel")

    plt.subplot(1, 4, 3)
    plt.imshow(G, cmap='Greens')
    plt.title("Green Channel")

    plt.subplot(1, 4, 4)
    plt.imshow(B, cmap='Blues')
    plt.title("Blue Channel")

    plt.show()

    return R, G, B
```

```python
def CountPix(I):
    """
    Returns the total number of pixels in the image.
    """
    return I.size

def FactPix(I, factor1, factor2):
    """
    Modifies the image intensity using two factors.
    - factor1 (float): Scaling factor (controls contrast), if > 1, increases contrast by making the dark pixls darker and the light ones lighter .
    - factor2 (float): Shift factor (controls brightness).
    the formula is: I' = factor1 * I + factor2
    """
    return cv2.convertScaleAbs(I, alpha=factor1, beta=factor2)

def Func_a(I):
    """
    Applies logarithm, exponential, square, and square root transformations.
    """
    log_I = np.log1p(I)
    exp_I = np.exp(I / 255.0) * 255
    sq_I = np.square(I)
    sqrt_I = np.sqrt(I)
    return log_I, exp_I, sq_I, sqrt_I

def Func_m(I):
    """
    Computes mean and standard deviation of the image.
    """
    mean, stddev = cv2.meanStdDev(I)
    return mean[0][0], stddev[0][0]

def Normalize(I, min_val, max_val):
    """
    Normalizes image intensity between min_val and max_val.
    """
    return cv2.normalize(I, None, min_val, max_val, cv2.NORM_MINMAX)

def Inverse(I):
    """
    Computes the negative of an image.
    """
    return 255 - I

def CalcHist(I):
    """
    Computes histogram of an image.
    """
    hist = cv2.calcHist([I], [0], None, [256], [0, 256])
    return hist

def Threshold(I, s):
    """
    Applies thresholding to an image.
    """
    _, T = cv2.threshold(I, s, 255, cv2.THRESH_BINARY)
    return T

def Func_j(image_path):
    """
    Opens an image, displays it, shows histogram, inverts it,
    and calculates histogram after inversion.
    """
    I, _, _ = OpenImage(image_path)
    plt.figure(figsize=(10, 5))
    plt.subplot(1, 3, 1)
    plt.imshow(I, cmap='gray')
    plt.title('Original')
    plt.subplot(1, 3, 2)
    plt.plot(CalcHist(I))
    plt.title('Histogram')
    Inv = Inverse(I)
    plt.subplot(1, 3, 3)
    plt.imshow(Inv, cmap='gray')
    plt.title('Inverted')
    plt.show()

def Func_t(image_path):
    """
    Opens an image, displays it, normalizes it between [10, 50],
    and shows its histogram.
    """
    I, _, _ = OpenImage(image_path)
    plt.figure(figsize=(10, 5))
    plt.subplot(1, 3, 1)
    plt.imshow(I, cmap='gray')
    plt.title('Original')
    Inorm = Normalize(I, 10, 50)
    plt.subplot(1, 3, 2)
    plt.imshow(Inorm, cmap='gray')
    plt.title('Normalized')
    plt.subplot(1, 3, 3)
    plt.plot(CalcHist(Inorm))
    plt.title('Histogram after Normalization')
    plt.show()

def Func_f(image_path):
    """
    Opens an image, applies thresholding at s=128,
    and displays the histogram before and after thresholding.
    """
    I, _, _ = OpenImage(image_path)
    plt.figure(figsize=(10, 5))
    plt.subplot(1, 3, 1)
    plt.imshow(I, cmap='gray')
    plt.title('Original')
    plt.subplot(1, 3, 2)
    plt.plot(CalcHist(I))
    plt.title('Histogram before Thresholding')
    T = Threshold(I, 128)
    plt.subplot(1, 3, 3)
    plt.imshow(T, cmap='gray')
    plt.title('Thresholded Image')
    plt.show()

# Exercise 2: Generate a 512x512 random matrix and visualize its histogram
def generate_random_matrix():
    """
    Creates a 512x512 random matrix using a normal distribution.
    """
    mean = 128
    stddev = np.random.randint(10, 50)
    I = np.random.normal(mean, stddev, (512, 512)).astype(np.uint8)
    plt.figure(figsize=(10, 5))
    plt.subplot(1, 2, 1)
    plt.imshow(I, cmap='gray')
    plt.title('Random Image')
    plt.subplot(1, 2, 2)
    plt.hist(I.ravel(), bins=256, range=[0, 256])
    plt.title('Histogram')
    plt.show()
    return I
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Fonction pour calculer l'histogramme
def CalcHist(I):
    H, bins = np.histogram(I.flatten(), bins=256, range=[0, 256])
    return H, bins[:-1]

# Définition de la matrice I (4x4)
I = np.array([
    [128, 128, 0, 255],
    [128, 0, 0, 255],
    [0, 128, 0, 255],
    [128, 128, 0, 255]
], dtype=np.uint8)

# Génération des matrices aléatoires (512x512)
ran_d = np.random.rand(512, 512) * 255  # Valeurs entre 0 et 255
ran_dN = np.random.randn(512, 512) * np.random.randint(10, 50) + 128  # Moyenne 128, std aléatoire
ran_dINT = np.random.randint(0, 256, size=(512, 512))  # Entiers entre 0 et 255

# Calcul des histogrammes
H_I, bins_I = CalcHist(I)
H_ran_d, bins_ran_d = CalcHist(ran_d)
H_ran_dN, bins_ran_dN = CalcHist(ran_dN)
H_ran_dINT, bins_ran_dINT = CalcHist(ran_dINT)

# Création d'une figure avec 4 lignes et 2 colonnes
fig, axes = plt.subplots(4, 2, figsize=(10, 12))

# 1. Affichage de Matrix I et son histogramme
axes[0, 0].imshow(I, cmap='gray')
axes[0, 0].set_title("Matrix I")
axes[0, 0].axis("off")

axes[0, 1].bar(bins_I, H_I, width=1, color='gray', alpha=0.7)
axes[0, 1].set_title("Histogram of Matrix I")

# 2. Affichage de Random (0-255) et son histogramme
axes[1, 0].imshow(ran_d, cmap='gray')
axes[1, 0].set_title("Random (0-255)")
axes[1, 0].axis("off")

axes[1, 1].bar(bins_ran_d, H_ran_d, width=1, color='gray', alpha=0.7)
axes[1, 1].set_title("Histogram of Random (0-255)")

# 3. Affichage de Normal Dist. et son histogramme
axes[2, 0].imshow(ran_dN, cmap='gray')
axes[2, 0].set_title("Normal Dist. (mean=128)")
axes[2, 0].axis("off")

axes[2, 1].bar(bins_ran_dN, H_ran_dN, width=1, color='gray', alpha=0.7)
axes[2, 1].set_title("Histogram of Normal Dist.")

# 4. Affichage de Random Integers et son histogramme
axes[3, 0].imshow(ran_dINT, cmap='gray')
axes[3, 0].set_title("Random Integers (0-255)")
axes[3, 0].axis("off")

axes[3, 1].bar(bins_ran_dINT, H_ran_dINT, width=1, color='gray', alpha=0.7)
axes[3, 1].set_title("Histogram of Random Integers")

# Ajustement et affichage
plt.tight_layout()
plt.show()
```
